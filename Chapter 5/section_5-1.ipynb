### Section 5.1: Systems of Linear Equations in Two Variables

##### Objective 1: Checking a Linear System for a Solution

##### Definitions:

A <em><b>linear equation of two variables</b></em> is an equation of the form $Ax + By = C$ for any real numbers $A, B,$ and $C$ (when plotted this forms a <em>straight line</em>).

A <em><b>system of linear equations</b></em> is a set of linear equations that must all be satisfied simultaneously.

##### Possible Solutions to a Linear System:

There are three general types of solutions to a linear sytem of equations:

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def f(x):
    return 2*x+3

def g(x):
    return -2*x+5

x = np.arange(-20,21,1)
y_f = f(x)
y_g = g(x)

fig, (ax1,ax2,ax3) = plt.subplots(3,figsize=(25,20))
ax1.plot(x,y_f,color='blue',label='$f(x)$')
ax1.plot(x,y_g,color='red',label='$g(x)$')
ax1.set_title('Unique Solution',fontsize=25)
ax1.grid()
ax1.set(xlim=(-10,10),ylim=(-10,10))
ax1.set_aspect('equal')
ax1.set_xticks(np.arange(-10,10,1))
ax1.set_yticks(np.arange(-10,10,1))
ax1.axhline(y=0,color='k')
ax1.axvline(x=0,color='k')
ax1.set_xlabel('$x$',fontsize=25)
ax1.set_ylabel('$y$',fontsize=25)
ax1.tick_params(labelsize=12)
ax1.legend()

ax2.plot(x,y_f,color='blue',label='$f(x)$')
ax2.plot(x,y_f+4,color='red',label='$g(x)$')
ax2.set_title('No Solution',fontsize=25)
ax2.grid()
ax2.set_aspect('equal')
ax2.set(xlim=(-10,10),ylim=(-10,10))
ax2.set_xticks(np.arange(-10,10,1))
ax2.set_yticks(np.arange(-10,10,1))
ax2.axhline(y=0,color='k')
ax2.axvline(x=0,color='k')
ax2.set_xlabel('$x$',fontsize=25)
ax2.set_ylabel('$y$',fontsize=25)
ax1.tick_params(labelsize=12)
ax2.legend()

ax3.plot(x,y_f,color='blue',label='$f(x)$')
ax3.plot(x,y_f,color='red',label='$g(x)$')
ax3.set_title('Infinitely Many Solutions',fontsize=25)
ax3.grid()
ax3.set_aspect('equal')
ax3.set(xlim=(-10,10),ylim=(-10,10))
ax3.set_xticks(np.arange(-10,10,1))
ax3.set_yticks(np.arange(-10,10,1))
ax3.axhline(y=0,color='k')
ax3.axvline(x=0,color='k')
ax3.set_xlabel('$x$',fontsize=25)
ax3.set_ylabel('$y$',fontsize=25)
ax1.tick_params(labelsize=12)
ax3.legend()

fig.tight_layout()
```

##### Definitions:

If a system of linear equations has no solution, then that system is <em><b>inconsistent</b></em>.

If a system of linear equations has at least one solution, then the system is <em><b>consistent</b></em>.

If a system of linear equations has infinitely many solutions, then the system is <em><b>dependent</b></em>.

##### Example:

Determine if the ordered pair $(1,-2)$ is a solution to the system: $$y = 3x-5 \\ 2x+6y = -10$$

##### Solution:

$\require{\cancel}$

We need to verify that the ordered pair satisfies <em>each</em> equation:

For the first equation we need to verify that $y = 3x -5$ is a true statement when $x = 1$ and $y = -2$:

$$y = 3(1) - 5 = 3 - 5 = -2 \, \checkmark$$

Likewise, we need to verify that the second equation, $2x + 6y = -10$ is also true when $x = 1$ and $y = -2$:

$$2(1) + 6(-2) = 2 - 12 = -10 \, \checkmark$$

So we can conclude that since <em>both</em> equations hold true for the given ordered pair, the system's solution is $(x,y)=(1,-2)$.

$$y = 3x - 5 \\ 2x + 6y = -10$$

We graph the two lines and then "eyeball" or essentially guess at the ordered pair where the lines meet:

We know from the previous example what the ordered pair should be. But we can also see from the graph that the ordered pair, $(1,-2)$, does appear to be where the two lines intersect.

So we would say that the solution is $(x,y) = (1,-2)$.

##### Objective 2: Solving Linear Systems by the Substitution Method

##### Steps for Solving a System of Equations of Two Variables Using Substitution:

1. Choose a variable ($x$ or $y$) and solve either of the equations for that variable.
2. Substitute or replace the variable in the other equation with the result obtained from step 1.
3. Solve the equation that results from the substitution in step 2 (there should be only a single variable to solve for at this stage).
4. Use the result from step 3 to substitute into <em>either</em> equation that you started with.
5. For a unique solution, the solution is then the ordered pair given by $(x,y)$. We will address "no solution" and "infinitely many solutions" later.

##### Example:

Solve the system of equations by the <em>substitution method</em>:

$$x+3y = 6 \\ y - 9 = 2x$$

##### Solution:

1. We first choose a variable. We'll solve for $y$ since the second equation looks setup well for that:

$$ y - \cancel{9} \color{blue}{+} \cancel{\color{blue}{9}} = 2x \color{blue}{+9} \Rightarrow y = \underline{2x + 9}$$

The underlined equation is the equation that results from solving the <em>2nd equation</em> for the variable $y$.

2. We can now use the fact that $y = \underline{2x + 9}$ in place of the variable $y$ in the <em>1st equation</em>:

$$x + 3y = 6 \Rightarrow x + 3(\color{blue}{2x+9}) = 6$$

3. Solving: 

$$x + \color{blue}{6x + 27} = 6$$

$$\color{blue}{7x} + 27 = 6$$

$$7x + \cancel{27} \color{blue}{-} \cancel{\color{blue}{27}} = 6 \color{blue}{-27}$$

$$7x = -21$$

$$\frac{\cancel{7}x}{\cancel{\color{blue}{7}}} = \frac{-21}{\color{blue}{7}} \Rightarrow \underline{x = -3}$$

So now we know that under the given conditions at the start, we <em>must</em> have $x = -3$ as part of our solution. We can use this knowledge to solve for the value of $y$.

If we go by the steps for solving via substitution given above, we could use this value for $x$ in either equation that we start with. However, we have already solved for $y$ in terms of $x$ from the 2nd equation; so we could just use that equivalent form of equation 2:

4. $$y = 2x + 9 \Rightarrow y = 2(\color{blue}{-3}) + 9 \Rightarrow y = 3$$

5. So $y = 3$ and we now have a value obtained for both $x$ and $y$. Therefore, our solution is the ordered pair given by $\boxed{(x,y) = (-3,3)}$.

##### Example:

Solve the system of equations by <em>substitution</em>:

$$5x + y = 9 \\ -x + 3y = 11$$

##### Solution:

1. Choose a variable:

We'll solve for $y$ in the first equation:

$$\cancel{5x} \color{blue}{-} \cancel{\color{blue}{5x}} + y = 9 \color{blue}{-5x} \Rightarrow y = \underline{9 - 5x}$$

2. Substituting:

$$-x + 3(\color{blue}{9-5x}) = 11$$

3. Solving:

$$-x + \color{blue}{27 - 15x} = 11$$

$$\color{blue}{-16x} + 27 = 11$$

$$-16x + \cancel{27} \color{blue}{-} \cancel{\color{blue}{27}} = 11 \color{blue}{-27}$$

$$-16x = -16$$

$$\frac{\cancel{-16}x}{\cancel{\color{blue}{-16}}} = \frac{-16}{\color{blue}{-16}}$$

$$\underline{x = 1}$$

4. Substituting this value into either equation (I'll use the equivalent form of equation 1):

$$\underline{y = 9 - 5(\color{blue}{1}) = 4}$$

5. So our solution is: $\boxed{(x,y)=(1,4)}$.

##### Objective 3: Solving Linear Systems by the Addition/Elimination Method

##### Steps for Solving a System of Equations of Two Variables Using Addition/Elimination:

1. Rewrite each equation in the form $Ax + By = C$ (if necessary).
2. Choose a variable to eliminate.
3. Multiply each equation by a number (if necessary) that will match the coefficients of the chosen variable in <em>both</em> equations with the exception that the coefficients are opposite in sign.
4. Add the two equations obtained from step 3.
5. Solve the resulting equation from step 4 for the variable that remains (the chosen variable in step 2 should be eliminated at this point).
6. Use the value obtained from step 5 to substitute into <em>either</em> starting equation.
7. The ordered pair obtained is the solution to the system.

##### Example:

Solve the system by the addition/elimination method:

$$2x + 6y = 6 \\ 4x - 6y = 12$$

##### Solution:

1. Both equations are already in the form $Ax + By = C$ (each variable is on the same side as the other with no other terms added/subtracted on that side).

2. Now we determine which variable to <em>eliminate</em>. Since the coefficients on the $y$ terms in <em>each</em> equation are already the same value but differ in sign only, $y$ would be the likely choice in this case.

3. Since both equations have the same coefficents for the $y$ terms and differ in sign already, we don't need to multiply either equation by anything. So we move on to step 4.

4. We add the two equations:

\begin{eqnarray}
\nonumber 2x + \cancel{6y} & = & 6 \\
\nonumber \color{blue}{+} \, 4x - \cancel{6y} & = & 12 \\
\hline
\nonumber \color{blue}{6x} & = & \color{blue}{18}
\end{eqnarray}

5. Now we solve:

$$\frac{\cancel{6}x}{\cancel{\color{blue}{6}}} = \frac{18}{\color{blue}{6}}$$

Therefore, $\underline{x = 3}$

6. Now we substitute this into either starting equation and solve. We'll use the first equation from the original problem statement:

$$2x + 6y = 6 \Rightarrow 2(\color{blue}{3}) + 6y = 6$$

Solving:

\begin{eqnarray}
\nonumber \color{blue}{6} + 6y = 6 & \Rightarrow & \cancel{6} \color{blue}{-} \cancel{\color{blue}{6}} + 6y = 6 \color{blue}{-6} \\
\nonumber & \Rightarrow & 6y = 0 \\
\nonumber & \Rightarrow & \frac{\cancel{6}y}{\cancel{\color{blue}{6}}} = \frac{0}{\color{blue}{6}} \\
\nonumber & \Rightarrow & \underline{y = 0}
\end{eqnarray}

7. Therefore, our solution is given by $\boxed{(x,y)=(3,0)}$.

##### Example:

Solve the system by the addition/elimination method:

$$2x+4y = 8 \\ 5x + 10y = 20$$

##### Solution:

1. Both equations are in the correct form.

2. We will choose to eliminate $x$.

##### Note:

We could just as easily choose to eliminate $y$. After examining the equation for a bit, neither variable appears to be "easier" to eliminate than the other. So it's really a matter of preference.

3. We now need to multiply each equation by a value so that the coefficients on both $x$ terms are the same in value, but opposite in sign:

$$\color{blue}{5\cdot}(2x+4y) = \color{blue}{5 \cdot} 8 \\ \color{blue}{-2 \cdot}(5x+10y) = \color{blue}{-2 \cdot} 20$$

This process is similar to finding a common denominator: we wish to find a value for which both coefficients are multiples. Here we have chosen $-2$ as the multiplier on equation 2, but we could have used $-5$ just as easily as the multiplier on equation 1. The important thing is that we end up with <em>opposite</em> signs for the $x$ terms.

We then simplify (distribute):

$$10x+20y = 40 \\ -10x-20y = -40$$

4. And now we add:

\begin{eqnarray}
\nonumber \cancel{10x} + \cancel{20y} & = & \cancel{40} \\
\nonumber \color{blue}{+} \, \cancel{-10x} - \cancel{20y} & = & \cancel{-40} \\
\hline
\nonumber \color{blue}{0} & = & \color{blue}{0}
\end{eqnarray}

5. Normally, we would solve this equation for the variable that remains, but we clearly have eliminated all variables! Since we have eliminated all variables and we have arrived at a true statement ($0 = 0$ is <em>always</em> true), we can declare that we have <em>infinitely many solutions</em>.

In fact, <em>every</em> ordered pair that satisfies the equation $2x+4y = 8$ will be the same set of ordered pairs that satisfy the equation $5x+10y = 20$, because they are the <em>same</em> equation, but in a different form! In other words, the two given equations are represented the <em>same line</em>, not two separate lines.

This also means we don't have a step 6 or 7. We can just immediate state the solution set has $\boxed{\text{infinitely many solutions}}$ and leave it at that.

##### Example:

Solve the system by addition method:

$$2x=y-7 \\ 2x-y=9$$

##### Solution:

1. The second equation is in the form $Ax+By=C$, but the first equation is not. So we need to do that first:

$$2x\color{blue}{-y}=\cancel{y} \color{blue}{-} \cancel{\color{blue}{y}}-7 \Rightarrow 2x - y = -7$$

So now our system looks like this:

$$2x-y = -7 \\ 2x-y=9$$

Now we could examine this system and notice that the left side of each equation is <em>exactly the same</em>, but they are equal to different values on the right. This suggests that this system will have no solution. But we'll go through and see what happens if we neglected this fact.

2. We choos a variable to eliminate. In this case, it really doesn't matter which one we choose, in both cases we just need the opposite signs.

3. So we multiply the second equation by $-1$:

$$2x - y = -7 \\ \color{blue}{-1 \cdot} (2x-y) = \color{blue}{-1 \cdot} 9$$

The second equation simplifies to $-2x + y = -9$.

4. So when we add:

\begin{eqnarray}
\nonumber \cancel{2x} - \cancel{y} & = & -7 \\
\nonumber \color{blue}{+} \, \cancel{-2x} + \cancel{y} & = & -9 \\
\hline
\nonumber \color{blue}{0} & \color{blue}{\neq} & \color{blue}{-16}
\end{eqnarray}

Since we now obtain a result that is a contradiction ($0=-16$ is <em>never</em> true), we can declare that there is no ordered pair that will satisfy the sytem $\Rightarrow$ there is $\boxed{\text{no solution}}$.

##### Objective 4: Applications of Solving Linear Systems

##### Example:

The sum of two numbers is $12$. If one number is subtracted from the other, the result is $-2$. Find the numbers.

##### Solution:

We can translate this scenario into a system of equations and then solve the system. We need to determine two equations:

The first sentence says the sum of two numbers (say $x$ and $y$) is $12 \Rightarrow x + y = 12$.

The second sentence says that if one number (again, using $x$ and $y$ as the numbers) is subtracted from the other, we obtain $-2 \Rightarrow x - y = -2$.

The system of equations is therefore given by $$x + y = 12 \\ x - y = -2$$

We could just add these two since the $y$ terms are setup to eliminate right away:

\begin{eqnarray}
\nonumber x + \cancel{y} & = & 12 \\
\nonumber \color{blue}{+} \, x - \cancel{y} & = & -2 \\
\hline
\nonumber \color{blue}{2x} & = & \color{blue}{10} \\
\nonumber \frac{\cancel{2}x}{\cancel{\color{blue}{2}}} & = & \frac{10}{\color{blue}{2}} \\
\nonumber x & = & 5
\end{eqnarray}

So we know that <em>one</em> of these two numbers is $\underline{x=5}$. We now use either of the first two equations to find the other number (using the first equation): $\color{blue}{5} + y = 12 \Rightarrow \underline{y = 12 - 5 = 7}$.

Therefore, the two numbers must be $\boxed{5 \text{ and } 7}$. If we add these numbers we do indeed get $12$ and if we subtract them (in the correct order) we will get $-2$.

##### Example:

Four times a first number decreased by a second number is $2$. The first number increased by twice the second number is $-13$. Find the numbers.

##### Solution:

Translating: we'll represent the first number as $x$ and the second number as $y$. This leads to the equations $4x - y = 2$ and $x + 2y = -13$.

So our system to be solved is given by $$4x-y=2 \\ x+2y = -13$$

We can eliminate either variable rather easily. Since the $y$ terms already are opposite signs of each other, we'll eliminate $y$:

$$\color{blue}{2\cdot} (4x - y) = \color{blue}{2 \cdot} 2 \\ x + 2y = -13$$

This leads to $$8x - 2y = 4 \\ x + 2y = -13$$

\begin{eqnarray}
\nonumber 8x - \cancel{2y} & = & 4 \\
\nonumber \color{blue}{+} \, x + \cancel{2y} & = & -13 \\
\hline
\nonumber \color{blue}{9x} & = & \color{blue}{-9} \\
\nonumber \frac{\cancel{9}x}{\cancel{\color{blue}{9}}} & = & \frac{-9}{\color{blue}{9}} \\
\nonumber x & = & -1
\end{eqnarray}

Using this value in the first equation:

$$4(\color{blue}{-1}) - y = 2 \Rightarrow -4 - y = 2$$

Adding $4$ to both sides:

$$-y = 6$$

Dividing both sides by $-1$:

$$y = -6$$

So the solutions set is $\boxed{-1 \text{ and } -6}$.
