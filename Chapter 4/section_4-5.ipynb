##### Objective 1: Modeling Exponential Growth and Decay

##### Recall Exponential Growth:

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def f(x):
    return np.exp(x)

xmin = 0
xmax = 5
xstep = 1

x = np.arange(xmin,xmax+.1,.1)
y = 50*f(x)

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x,y)
ax.axhline(y=0,color='k')
ax.axvline(x=0,color='k')
ax.set_title('Exponential Growth: $A = A_0e^{kt}$, where $k > 0$',fontsize=25)
ax.set_xticks(np.arange(xmin,xmax+1,1))
ax.set_yticks(np.arange(0,max(y)+1,1000))
ax.set_xlabel('$t$',fontsize=25)
ax.set_ylabel('$A$',fontsize=25)
ax.tick_params(labelsize=25)
ax.grid()
```

##### Recall Exponential Decay:

```python
xmin = 0
xmax = 5
xstep = 1

x = np.arange(xmin,xmax+.1,.1)
y = 50*f(-x)

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x,y)
ax.axhline(y=0,color='k')
ax.axvline(x=0,color='k')
ax.set_title('Exponential Decay: $A=A_0e^{kt}$, where $k < 0$',fontsize=25)
ax.set_xticks(np.arange(xmin,xmax+1,1))
ax.set_yticks(np.arange(0,max(y)+1,10))
ax.set_xlabel('$t$',fontsize=25)
ax.set_ylabel('$A$',fontsize=25)
ax.tick_params(labelsize=25)
ax.grid()
```

##### Example:

The exponential model $A = 27.5e^{0.003t}$ describes the population, $A$, of a country in millions, $t$ years after 2003. Use the model to determine when the population of the country will be 30 million (round to the nearest whole year).

##### Solution:

What we know:
* $A_0 = 27.5$
* $k = 0.003$
* Looking for a total amount of 30 million $\Rightarrow A = 30$

So we have $30 = 27.5e^{0.003t}$. The only thing we don't know is $t$ (number of years since 2003).

Now we solve:

\begin{eqnarray}
30 & = & 27.5e^{0.003t} \\
\frac{30}{\color{blue}{27.5}} & = & \frac{27.5e^{0.003t}}{\color{blue}{27.5}} \\
\color{blue}{\ln} \frac{30}{27.5} & = & \color{blue}{\ln} e^{0.003t} \\
\ln \frac{30}{27.5} & = & \color{blue}{0.003t} \\
\frac{\ln \left( \frac{30}{27.5} \right)}{\color{blue}{0.003}} & = & \frac{0.003t}{\color{blue}{0.003}} \\
t & \approx & 29
\end{eqnarray}

Therefore, about 29 years <em>after</em> the year 2003 will be when the country will reach 30 million. So $2003+29 = 2032 \Rightarrow$ $\boxed{\text{in the year 2032}}$.

##### Example:

Suppost an artifact has 16 grams of carbon-14 present. The decay model $A = 16e^{-0.000121t}$ describes the amount of carbon-14 present after $t$ years. Use the model to determine how many grams of carbon-14 will be present in 5418 years (round to the nearest whole gram).

##### Solution:

The decay model is given: $A = 16e^{-0.000121t}$

Here, we can just "plug-in" $t=5418$ in the decay model.

$A = 16e^{-0.000121 \cdot 5418} \approx 8 \Rightarrow$ $\boxed{\text{about } 8 \text{ grams of carbon-14 will be present } 5418 \text{ years from now}}$ (about half the original amount of carbon-14).

##### Definition:

The <em><b>half-life</b></em>, $t_{1/2}$, is the time it takes for some initial quantity to reach <em>half</em> its original amount.

So in the previous example, since about half of the original carbon-14 is present after 5418 years, we would say that $t_{1/2} \approx 5418$ years.

##### Example:

Prehistoric cave paintings were discovered in a cave in France. The paint contained 6\% of the original carbon-14. Use the exponential decay model for carbon-14, $A=A_0e^{-0.000121t}$, to estimate the age of the paintings (round to the nearest whole year).

##### Solution:

We're given the decay model for carbon-14: $A = A_0e^{-0.000121t}$. We are not given an initial amount for $A_0$.

What we are given is the <em>remaining percentage</em> of carbon-14: $A =$ 6\% of the original amount.

Or in symbols: $A = 0.06A_0$.

Since we know $A = A_0 e^{-0.000121t}$ and we also know that $A = 0.06A_0$, then $0.06A_0 = A_0 e^{-0.000121t}$.

So we solve this equation:

First, note that $A_0 \ne 0$:

\begin{eqnarray}
0.06A_0 & = & A_0 e^{-0.000121t} \\
\frac{0.06A_0}{\color{blue}{A_0}} & = & \frac{A_0 e^{-0.000121t}}{\color{blue}{A_0}} \\
0.06 & = & e^{-0.000121t} \\
\color{blue}{\ln} 0.06 & = & \color{blue}{\ln} e^{-0.000121t} \\
\ln 0.06 & = & \color{blue}{-0.000121t} \\
\frac{\ln 0.06}{\color{blue}{-0.000121}} & = & \frac{-0.000121t}{\color{blue}{-0.000121}} \\
t & \approx & 23251
\end{eqnarray}

So the cave painting is about 23,251 years old.

So the cave painting is about $\boxed{23,251 \text{ years old}}$.

##### Example:

Complete the table for the half-life of a certain radioactive substance.

<table style="width:75%">
    <tr> <th>Half-life</th> <th>Decay Rate, $k$</th> </tr>
    <tr> <td></td> <td>6.7% per year</td> </tr>
</table>

##### Solution:

We know <em>two</em> things to start with: 1) we require the decay model $(A = A_0e^{kt})$ and 2) we can determine from the table that $k = -0.067$.

For finding $k$ in part 2) above, we know that $k$ should be <em>negative</em> since we are dealing with a <em>decay rate</em> and we convert the percentage into a decimal.

With both of these knowns together we have $A = A_0e^{-0.067t}$ as our decay model.

We need to use this decay model to find $t_{1/2}$. Since, by definition, $t_{1/2}$ means that $A = $ 50\% of the original amount, we require that $A = 0.50A_0$.

This means that we need to solve $0.5A_0 = A_0 e^{-0.067t}$ for $t$:

\begin{eqnarray}
0.50A_0 & = & A_0 e^{-0.067t} \\
\frac{0.50A_0}{\color{blue}{A_0}} & = & \frac{A_0 e^{-0.067t}}{\color{blue}{A_0}} \\
0.50 & = & e^{-0.067t} \\
\color{blue}{\ln} 0.50 & = & \color{blue}{\ln} e^{-0.067t} \\
\ln 0.50 & = & \color{blue}{-0.067t} \\
\frac{\ln 0.50}{\color{blue}{-0.067}} & = & \frac{-0.067t}{\color{blue}{-0.067}} \\
t & \approx & 10
\end{eqnarray}

So that means $\boxed{t_{1/2} \approx 10 \text{ years}}$.

##### Example:

The half-life of thorium-229 is about 7340 years. About how long will it take for a sample of this substance to decay to 20\% of its original amount?

##### Solution:

We are dealing with another half-life scenario $\Rightarrow$ use exponential decay model: $A = A_0 e^{kt}$.

We know what the half-life is: $t_{1/2} = 7340$ years.

What we desire to know: total time it takes for the substance to have 20\% of its original amount.

If we just focus on what we know first:

We know that we need to use, $A = A_0 e^{kt}$, but here we don't know what $k$ is! So we'll need to find that information somehow.

We also know that $t_{1/2} = 7340$ is the <em>half-life</em> for this substance. By definition of half-life $\Rightarrow A = 0.50A_0$ when $t = t_{1/2} = 7340$ (this is true no matter what $k$ is).

Therefore, we can say $0.50A_0 = A_0 e^{k \cdot (7340)}$ must be true for some value $k$. And now we can solve for $k$. So let's go ahead and do that:

\begin{eqnarray}
0.50A_0 & = & A_0 e^{7340k} \\
\frac{0.50A_0}{\color{blue}{A_0}} & = & \frac{A_0 e^{7340k}}{\color{blue}{A_0}} \\
\color{blue}{\ln} 0.50 & = & \color{blue}{\ln} e^{7340k} \\
\ln 0.50 & = & \color{blue}{7340k} \\
\frac{\ln 0.50}{\color{blue}{7340}} & = & \frac{7340k}{\color{blue}{7340}} \\
k & \approx & -0.000094
\end{eqnarray}

##### Note:

When solving for $k$ in either growth or decay models, it's a good idea to round to at least six decimal places to avoid rounding error in calculations.

Now that we know what $k$ is we can now investigate how long it takes for this substance to decay down to 20\% of its original amount.

So we go back to the decay model, but with an unknown time, $t$, and with our known $k$ value:

$A = A_0 e^{-0.000094t}$

And now our problem looks very much like the previous problems involving carbon dating. The difference being in the carbon dating problems we were given a $k$ value. And in this case, since we are dealing with a different substance, we had to find the correct $k$ value and then proceed.

For thorium-229 to decay to 20\% of its original amount then:

\begin{eqnarray}
0.20A_0 & = & A_0 e^{-0.000094t} \\
\frac{0.20A_0}{\color{blue}{A_0}} & = & \frac{A_0 e^{-0.000094t}}{\color{blue}{A_0}} \\
\color{blue}{\ln} 0.20 & = & \color{blue}{\ln} e^{-0.000094t} \\
\ln 0.20 & = & \color{blue}{-0.000094t} \\
\frac{\ln 0.20}{\color{blue}{-0.000094}} & = & \frac{-0.000094t}{\color{blue}{-0.000094}} \\
t & \approx & 17122
\end{eqnarray}

So it'll take about $\boxed{t = 17,122 \text{ years}}$ for any amount of thorium-229 to decay to 20\% of its original amount.
