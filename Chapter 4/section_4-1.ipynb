##### Obejctive 1: Graphing Exponential Functions

##### Exponential Function

* An <em><b>exponential function</b></em> is a function of the form $f(x) = b^{x}$, where b is a <em>positive</em> constant ($b > 0$)

##### Example 1

Graph $f(x) = 3^{x}$.

##### Solution:

* Choose some convenient $x$ values and evaluate:

<table style="width:50%">
    <tr> <th> $x$ </th> <th> $y = f(x)$ </th> </tr>
    <tr> <td> $-2$ </td> <td> $3^{-2} = \frac{1}{9}$ </td> </tr>
    <tr> <td> $-1$ </td> <td> $3^{-1} = \frac{1}{3}$ </td> </tr>
    <tr> <td> $0$ </td> <td> $3^0=1$ </td> </tr>
    <tr> <td> $1$ </td> <td> $3^1=3$ </td> </tr>
    <tr> <td> $2$ </td> <td> $3^2=9$ </td> </tr>
</table>

* Plotting these points in a plane:

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
def f(x):
    return 3**x

x = np.arange(-3,2.2,.1)
y = f(x)

x_sample = np.array([-2,-1,0,1,2])
y_sample = np.array([1/9,1/3,1,3,9])

fig, ax = plt.subplots(figsize=(25, 15))
ax.scatter(x_sample, y_sample, s=200, color='b')
ax.plot(x, y, color='b')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_xticks(np.arange(-2, 3))
ax.set_xticklabels(np.arange(-2,3), fontsize=25)
ax.set_yticks(np.arange(-1,11))
ax.set_yticklabels(np.arange(-1,11), fontsize=25)
ax.set_title('f(x)', fontsize=25)
```

* The line $y = 0$ ($x$-axis) is a <em><b>horizontal asymptote</b></em>. All exponential funtions have a horizontal asypmtote

* The <em><b>domain</b></em> (set of input) of the function (any exponential function) is $(-\infty, \infty)$ and the <em><b>range</b></em> (set of output) of the function is $(0,\infty)$.

##### Example 2

Graph $g(x) = \left( \frac{1}{3} \right)^{x}$.

##### Solution:

* First, we identify some points on the graph (choose some $x$ values and plug them in):

<table style="width:50%">
    <tr> <th> $x$ </th> <th> $y = g(x)$ </th> </tr>
    <tr> <td> $-2$ </td> <td> $\left(\frac{1}{3}\right)^{-2} = 9$ </td> </tr>
    <tr> <td> $-1$ </td> <td> $\left(\frac{1}{3}\right)^{-1} = 3$ </td> </tr>
    <tr> <td> $0$ </td> <td> $\left(\frac{1}{3}\right)^0=1$ </td> </tr>
    <tr> <td> $1$ </td> <td> $\left(\frac{1}{3}\right)^1=\frac{1}{3}$ </td> </tr>
    <tr> <td> $2$ </td> <td> $\left(\frac{1}{3}\right)^2=\frac{1}{9}$ </td> </tr>
</table>

* What similarities does this table have with the table from Example 1?

* The $y$ values are reversed! So perhaps we can rewrite $g(x)$ using notation involving $f(x)$?

* Say, $f(-x)$?

* If we consider $f(x) = 3^{x}$ from the previous example, then $f(-x) = 3^{-x}$. If we use the same choices for $x$ and look at $f(-x)$ in a table we get:

<table style="width:50%">
    <tr> <th> $x$ </th> <th> $y = g(x)$ </th> </tr>
    <tr> <td> $-2$ </td> <td> $3^{-(-2)} = 9$ </td> </tr>
    <tr> <td> $-1$ </td> <td> $3^{-(-1)} = 3$ </td> </tr>
    <tr> <td> $0$ </td> <td> $3^{-(0)}=1$ </td> </tr>
    <tr> <td> $1$ </td> <td> $3^{-1}=\frac{1}{3}$ </td> </tr>
    <tr> <td> $2$ </td> <td> $3^{-2}=\frac{1}{9}$ </td> </tr>
</table>

* So if we have a base, $b$, that is a fraction, we can flip the fraction and toss a negative in front of the exponent, $x$. That is, if $f(x) = \left( \frac{n}{m} \right)^{x}$, then $f(x) = \left( \frac{m}{n} \right)^{-x}$. In particular, for these first two examples, $g(x) = \left( \frac{1}{3} \right)^{x} = \left( \frac{3}{1} \right)^{-x} = 3^{-x} = f(-x)$.

* Plotting $g(x)$ then looks like the following.

```python
def g(x):
    return 3**(-x)

x = np.arange(-2.1,2.2,.1)
y = g(x)

y_parent = f(x)

x_sample = np.array([-2,-1,0,1,2])
y_sample = np.array([9,3,1,1/3,1/9])

fig, ax = plt.subplots(figsize=(25, 15))
ax.scatter(x_sample, y_sample, s=200)
ax.plot(x, y_parent, color='red', label=r'$f(x)$')
ax.plot(x, y, color='blue', label='$g(x)$')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_xticks(np.arange(-2, 3))
ax.set_xticklabels(np.arange(-2,3), fontsize=25)
ax.set_yticks(np.arange(-1,11))
ax.set_yticklabels(np.arange(-1,11), fontsize=25)
ax.set_title('g(x)', fontsize=25)
ax.legend(fontsize=25)
```

* As we can see here, the graph of $g(x)$ is reflected about the $y$ axis, just as should happen to a function $f(x)$ with the transformation $f(-x)$.

* The <em><b>domain</b></em> of $g(x)$ is $(-\infty,\infty)$.

* The <em><b>range</b></em> of $g(x)$ is $(0,\infty)$.

* The <em><b>horizontal asymptote</b></em> of $g(x)$ is $y = 0$.

##### Example 3

Graph $g(x) = 3 ^{x+4}$.

##### Solution:

* We could plot individual points... or we could use our knowledge about transformations of functions

* In particular, if $f(x) = 3^{x}$, then $f(x+4) = 3^{x+4} = g(x)$

* This means if we look at the graph of $f(x)$ as we had in the first example, we could just "shift" that graph $4$ units to the <em>left</em> (opposite sign in this case)

```python
def g(x):
    return 3**(x+4)

x = np.arange(-10,10.2,.1)
y = g(x)

y_parent = f(x)

fig, ax = plt.subplots(figsize=(25, 15))
ax.plot(x,y_parent,color='red', label=r'$f(x)$')
ax.plot(x, y,color='blue', label=r'$g(x)$')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_xticks(np.arange(-10, 11))
ax.set_xticklabels(np.arange(-10, 11), fontsize=25)
ax.set_xlim(-10, 10)
ax.set_yticks(np.arange(0, 11))
ax.set_yticklabels(np.arange(0, 11), fontsize=25)
ax.set_ylim(-1, 10)
ax.set_title('g(x)', fontsize=25)
ax.legend(fontsize=25)
```

* <em><b>Domain</b></em>: $(-\infty, \infty)$

* <em><b>Range</b></em>: $(0,\infty)$

* <em><b>Horizontal asymptote</b></em>: $y = 0$

##### Example 4

Graph $g(x) = 3^{\displaystyle x}+4$.

##### Solution:

* We need to shift the parent function's graph $4$ units up

* This, in effect, will also shift the horizontal asymptote of the parent function up by $4$ units

```python
def g(x):
    return 3**x+4

x = np.arange(-10,10.2,.1)
y = g(x)

fig, ax = plt.subplots(figsize=(25, 15))
ax.plot(x,y_parent,color='red', label=r'$f(x)$')
ax.plot(x, y,color='blue', label=r'$g(x)$')
ax.axhline(y=0, color='k')
ax.axhline(y=4, color='k',linestyle='--')
ax.axvline(x=0, color='k')
ax.set_xticks(np.arange(-10, 11))
ax.set_xticklabels(np.arange(-10, 11), fontsize=25)
ax.set_xlim(-10, 10)
ax.set_yticks(np.arange(0, 11))
ax.set_yticklabels(np.arange(0, 11), fontsize=25)
ax.set_ylim(-1, 10)
ax.set_title('g(x)')
ax.legend(fontsize=25)
```

* <em><b>Domain</b></em>: $(-\infty,\infty)$

* <em><b>Range</b></em>: $(4,\infty)$

* <em><b>Horizontal asymptote</b></em>: $y = 4$

##### Example 5

Graph $g(x) = -2^{-x+1} - 3$.

##### Solution:

* Factoring the exponent: $g(x) = -2^{-(x-1)} - 3$

* From factoring the negative in the exponent we now conclude we need a reflection over $x$-axis; reflection over $y$-axis; shift 1 unit to the right; and shift 3 units down.

```python
def g(x):
    return -2**(-x+1)-3

def f(x):
    return 2**x

x = np.arange(-10, 10.2, .1)
y = g(x)
y_parent = f(x)

fig, ax = plt.subplots(figsize=(25, 15))
ax.plot(x,y_parent,color='red', label=r'$f(x)$')
ax.plot(x, y,color='blue', label=r'$g(x)$')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axhline(y=-3, color='k',linestyle='--')
ax.axvline(x=0, color='k')
ax.set_xticks(np.arange(-10, 11))
ax.set_xticklabels(np.arange(-10, 11), fontsize=25)
ax.set_xlim(-10, 10)
ax.set_yticks(np.arange(-10, 11))
ax.set_yticklabels(np.arange(-10, 11), fontsize=25)
ax.set_ylim(-10, 10)
ax.set_title('Graph f(x)')
ax.legend(fontsize=25)
```

* <em><b>Domain</b></em>: $(-\infty,\infty)$

* <em><b>Range</b></em>: $(-\infty, -3)$

* <em><b>Horizontal asymptote</b></em>: $y = -3$

##### Objective 3: Applications of Exponential Functions

##### Compound Interest

* One area that makes extensive use of exponential functions is <em><b>compound interest</b></em>.

* Compound interest is interest that is computed on an orignal investment/loan as well as on any accumulated interest.

* The formula for finding the total amount accrued after some time period, $t$, is found by $A = P\left( 1 + \frac{r}{n} \right)^{nt}$.

* In the formula we have the following definitions:

    * $A$ is the future amount or face value.

    * $P$ is the principal or starting amount.

    * $r$ is the interest rate (expressed as a <em>decimal</em>).

    * $t$ is the time (in years) that has passed from the time of the principal investment/loan.

    * $n$ is the number of times per year the amount is compounded (number of times per year the interest is applied).

* Common values for $n$ are determined by key phrases:

    * Annually (once per year): $n= 1$.

    * Semiannually (twice per yer): $n = 2$.

    * Quarterly (four times per year): $n = 4$.

    * Monthly (12 months per year): $n = 12$.

    * Daily (365 days per year, not accounting for Leap Year): $n = 365$.

##### Continuous compounding interest formula:

* As n becomes very, very large, without bound, the future amount levels off

* This is what happens when we have <em>continous</em> compounding

* That is, when we decide to push $n$ to infinity (compounding infinitely many times per year, so to speak)

* When we do this, the original compound interest formula approaches the formula:

    * $A = Pe^{rt}$

        * $A$ is the future amount.

        * $P$ is the principal amount.

        * $r$ is the interest rate (again, in decimal form).

        * $t$ is time (in years).

##### Example 7

You have $\$1000$ to invest over a $3$-year period. You have the choice to invest at $8\%$ interest compounded daily; or at $6\%$ interest compounded continuously. Which choice will yield a larger future amount?

##### Solution:

* First we identify what we know so far. We are investing an initial amount of $P=\$1000$

* We are investing over a time period of $t=3$ years

* **Choice a**:
    
    * Interest rate, $r = 8\% = 0.08$ (divide $8\%$ by $100$)
    
    * We are compounding the interest <em>daily</em>, which means $n = 365$
    
    * We are not compounding the interest <em>continuously</em>, so we need to use the formula $A = P\left( 1 + \frac{r}{n} \right)^{nt}$

* **Choice b**:

    * Interest rate, $r = 6\% = 0.06$
    
    * We are compounding continuously so we need to use the formula $A = Pe^{rt}$

* Solving each case side-by-side:

<table style="width:75%">
    <tr> <th>Choice a</th> <th>Choice b</th> </tr>
    <tr> <td> $A = P\left( 1 + \frac{r}{n} \right)^{nt}$ </td> <td> $A = Pe^{rt}$ </td> </tr>
    <tr> <td> $A = 1000 \left( 1 + \frac{0.08}{365} \right)^{(365)(3)}$ </td> <td> $A = 1000e^{(0.06)(3)}$ </td> </tr>
    <tr> <td> $A \approx \$1271.22$ </td> <td> $A \approx \$1197.22$ </td> </tr>
</table>

* Therefore, choice a seems to be the better choice since we wish to maximize our investment

##### Example 8

You have $\$3000$ to invest over a $5$-year period. You have the choice to invest at $8.75\%$ interest compounded quarterly; or at $10\%$ compounded <em>continuously</em>. Which choice will yield a larger future amount?

##### Solution:

* Similar to the previous example we have two choices

* **Choice a**:

    * Interest rate, $r = 8.75\% = 0.0875$
    
    * We are compounding the interest <em>quarterly</em>, which means $n = 4$
    
    * We are not compounding the interest <em>continuously</em>, so we need to use the formula $A = P\left( 1 + \frac{r}{n} \right)^{nt}$

* **Choice b**:

    * Interest rate, $r = 10\% = 0.10$
    
    * We are compounding <em>continuously</em> so we need to use the formula $A = Pe^{rt}$

* Both choices involve an initial investment of $P=\$3000$ and a time period of $t=5$ years

* Solving each case side-by-side:

<table style="width:75%">
    <tr> <th>Choice a</th> <th>Choice b</th> </tr>
    <tr> <td> $A = P\left( 1 + \frac{m}{n} \right)^{nt}$ </td> <td> $A = Pe^{rt}$ </td> </tr>
    <tr> <td> $A = 3000 \left( 1 + \frac{0.0875}{4} \right)^{(4)(5)}$ </td> <td> $A = 1000 e^{(0.10)(5)}$ </td> </tr>
    <tr> <td> $A \approx \$4624.63$ </td> <td> $A \approx \$4946.16$ </td> </tr>
</table>

* Therefore, choice b seems to be the better choice since we wish to maximize our investment.
