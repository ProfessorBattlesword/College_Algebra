##### Objective 1: Identify intervals on which a function increases, decreases, or is constant

##### Increasing, Decreasing, and Constant

* A function is **increasing** on an open interval, $I$, if $f(x_1) < f(x_2)$, whenever $x_1 < x_2$ for any $x_1$ and $x_2$ in the interval

* A function is **decreasing** on an open interval, $I$, if $f(x_1) > f(x_2)$, whenever $x_1 < x_2$  for any $x_1$ and $x_2$ in the interval

* A functtion is **constant** on an open interval, $I$, if $f(x_1) = f(x_2)$, for any $x_1$ and $x_2$ in the interval

##### Objective 2: Use graphs to locate relative maxima or minima

##### Relative Minima and Maxima

* A functiuon value $f(a)$ is a **relative maximum (plural: relative maxima)** of $f$ if there exists an open inteval containing $a$ such that $f(a) > f(x)$ for all $x \neq a$ in the open interval

* A functiuon value $f(a)$ is a **relative minimum (plural: relative minima)** of $f$ if there exists an open inteval containing $a$ such that $f(a) < f(x)$ for all $x \neq a$ in the open interval

##### Objective 3: Identify even or odd functions and recognize their symmetries

##### Even and Odd Functions

* The function $f$ is an **even function** if $f(-x) = f(x)$ for all $x$ in the domain of $f$

    * The graph of an **even function** is *symmetric with respect to the $y$-axis*
    
* The function $f$ is an **odd function** if $f(-x) = -f(x)$ for all $x$ in the domain of $f$

    * The graph of an **odd function** is *symmetric with respect to the origin*

* If a function $f$ has neither property as described above, then $f$ is *neither* even or odd

##### Example 5:

Determine if the following functions are even, odd, or neither.

**(a)** $f(x) = x^5 + x^3$

**(b)** $f(x) = x^4 + 3$

**(c)** $f(x) = x^3 + x^2$

##### Solution:

**(a)**

\begin{align}
    f(-x) & = (-x)^5 + (-x)^3 \\
    & = -x^5 - x^3 \\
    & = -(x^5 + x^3) \\
    & = -f(x) \\
    & \Downarrow \\
    & \text{$f$ is odd}
\end{align}

**(b)**

\begin{align}
    f(-x) & = (-x)^4 + 3 \\
    & = x^4 + 3 \\
    & = f(x) \\
    & \Downarrow \\
    & \text{$f$ is even}
\end{align}

**(c)**

\begin{align}
    f(-x) & = (-x)^3 + (-x)^2 \\
    & = -x^3 + x^2 \\
    & \Downarrow \\
    f(-x) & \neq f(x) \\
    & \text{and} \\
    f(-x) & \neq -f(x) \\
    & \Downarrow \\
    & \text{$f$ is neither even or odd}
\end{align}

##### Objective 4: Understand and use piecewise functions

##### Piecewise Functions

* A function that is defined by two or more equations over a specified domain is called a **piecewise function**

##### Example 6:

Evaluate the piecewise-defined function below at the indicated values.

$$
    f(x) = \begin{cases} 2 - x & \text{if } x \geq -6 \\ x^2 & \text{if } x < -6 \end{cases}
$$

**(a)** $f(-10)$

**(b)** $f(-7)$

**(c)** $f(-6)$

**(d)** $f(4)$

##### Solution:

**(a)** $f(-10)$

* First we note that we are evaluating the function at $x = -10$ so we need to check the conditions to the right of each rule in the piecewise definition

* Since $x = -10$ satisfies the condition $x < -6$ (since $-10$ is less than $-6$) we use the formual to its left:

\begin{align}
    \text{Using the formula } \color{blue}{x^2}: f(-10) & = (-10)^2 \\
    & = \boxed{100}
\end{align}

**(b)** $f(-7)$

* We note that we are evaluating the function at $x = -7$ and check the conditions

* $x = -7$ also satisfies the condition $x < -6$ so we use the same formula as before:

\begin{align}
    \text{Using the formula } \color{blue}{x^2}: f(-7) & = (-7)^2 \\
    & = \boxed{49}
\end{align}

**(c)** $f(-6)$

* We note that we are evaluating the function at $x = -6$ and check the conditions

* $x = -6$ satisfies the condition $x \geq -6$ so we use the formula to its left:

\begin{align}
    \text{Using the formula } \color{blue}{2-x}: f(-6) & = 2 - (-6) \\
    & = 2 + 6 \\
    & = \boxed{8}
\end{align}

**(d)** $f(4)$

* We note that we are evaluating the function at $x = 4$ and check the conditions

* $x = 4$ satisfies the condition $x \geq -6$ so we use the formula to its left:

\begin{align}
    \text{Using the formula } \color{blue}{2 - x}: f(4) & = 2 - 4 \\
    & = \boxed{-2}
\end{align}

##### Example 8:

Graph the piecewise-defined function and determine the domain and range of the function.

$$
    f(x) = \begin{cases} 4x - 4 & \text{if } x < 2 \\ -x + 1 & \text{if } x \geq 2 \end{cases}
$$

##### Solution:

* We might notice that both formulas that appear in the piecewise function $f(x)$ represent two different lines in the form of $y = mx + b$

* To graph the piecewise function we then are just graphing lines over the portion of the domain that satisfies the condition for $x$

* That is, whenever $x < 2$ we have the graph of the line $y = 4x - 4$ and whenever $x \geq 2$ we have the graph of the line $y = -x + 1$

```python
fig, ax = plt.subplots(figsize=(25,15))

x_1 = np.linspace(-10, 1.98, 100)
x_2 = np.linspace(2.02, 10, 100)
y_1 = 4*x_1 - 4
y_2 = -x_2 + 1
ax.plot(x_1, y_1, color='b')
ax.plot(x_2, y_2, color='b')
open_func = patches.Circle((2, 4), radius=0.1, edgecolor='b',
                            fill=False, lw=2)
closed_func = patches.Circle((2, -1), radius=0.1, color='b')
ax.add_artist(open_func)
ax.add_artist(closed_func)
ax.axvline(x=0, color='k')
ax.axhline(y=0, color='k')
ax.set_xlim(-6,6)
ax.set_ylim(-6, 6)
ax.set_xticks(np.arange(-6, 8, 1))
ax.set_yticks(np.arange(-6, 7, 1))
ax.tick_params(labelsize=20)
ax.set_title(r'$f(x)$', fontsize=30)
ax.set_aspect('equal')
```

Domain: $(-\infty, \infty)$

Range: $(-\infty, 4)$

##### Objective 5: Find and simplify a function's difference quotient

##### The Difference Quotient

* The **difference quotient** of the function $f$ is defined by
$$
    \text{ } \\
    \frac{f(x + h) - f(x)}{h}, h \neq 0
$$

    * The difference quotient is one of the fundamental formulas behind the concept of derivative in calculus and you will see it again if you take calculus

##### Example 11:

Find the difference quotient of $f$ $\left( \text{i.e. find } \frac{f(x + h) - f(x)}{h}, h \neq 0 \right)$ for the following functions:

**(a)** $f(x) = 6x + 1$

**(b)** $f(x) = x^2 + 2x - 3$

##### Solution:

**(a)** $f(x) = 6x + 1$

* First, we find $f(x + h)$:

\begin{align}
    f(\color{blue}{x + h}) & = 6\color{blue}{(x + h)} + 1 \\
    & = \color{blue}{6x + 6h} + 1 \\
    f(x + h) & = 6x + 6h + 1
\end{align}

* Next, we compute $f(x + h)- f(x)$:

\begin{align}
    \color{red}{f(x + h)} - \color{orange}{f(x)} & = \color{red}{(6x + 6h + 1)} - \color{orange}{(6x + 1)} \\
    & = 6x + 6h + 1 \color{blue}{- 1 \cdot} (6x + 1) \\
    & = 6x + 6h + 1 \color{blue}{-6x - 1} \\
    & = \require{cancel} \cancel{6x} + 6h \cancel{+ 1} \cancel{-6x} \cancel{- 1} \\
    & = 6h
\end{align}

* Finally, we compute $\displaystyle \frac{f(x + h) - f(x)}{h}$:

\begin{align}
    \frac{\color{blue}{f(x + h) - f(x)}}{h} & = \frac{\color{blue}{6h}}{h} \\
    & = \frac{6\cancel{h}}{\cancel{h}} \\
    \frac{f(x + h) - f(x)}{h} & = \boxed{6}
\end{align}

**(b)** $f(x) = x^2 + 2x - 3$

* First, we find $f(x + h)$:

\begin{align}
    f(\color{blue}{x + h}) & = \color{blue}{(x + h)}^2 + 2 \color{blue}{(x + h)} - 3 \\
    & = (x + h)^2 + \color{blue}{2x + 2h} - 3 \\
    & = \color{blue}{(x + h)(x + h)} + 2x + 2h - 3 \\
    & = \color{blue}{x^2 + 2hx + h^2} + 2x + 2h - 3 \\
\end{align}

* Next, we compute $f(x + h) - f(x)$:

\begin{align}
    \color{red}{f(x + h)} - \color{orange}{f(x)} & = \color{red}{(x^2 + 2hx + h^2 + 2x + 2h - 3)} - \color{orange}{(x^2 + 2x - 3)} \\
    & = x^2 + 2hx + h^2 + 2x + 2h - 3 \color{blue}{- 1 \cdot} (x^2 + 2x - 3) \\
    & = x^2 + 2hx + h^2 + 2x + 2h - 3 \color{blue}{- x^2 - 2x + 3} \\
    & = \cancel{x^2} + 2hx + h^2 \cancel{+ 2x} + 2h \cancel{- 3} \cancel{- x^2} \cancel{- 2x} \cancel{+ 3} \\
    f(x + h) - f(x) & = 2hx + h^2 + 2h
\end{align}

* Finally, we compute $\displaystyle \frac{f(x + h) - f(x)}{h}$:

\begin{align}
    \frac{\color{blue}{f(x + h) - f(x)}}{h} & = \frac{\color{blue}{2hx + h^2 + 2h}}{h} \\
    & = \frac{\color{blue}{h \cdot (2x + h + 2)}}{h} \\
    & = \frac{\cancel{h}(2x + h + 2)}{\cancel{h}} \\
    \frac{f(x + h) - f(x)}{h} & = \boxed{2x + h + 2}
\end{align}
