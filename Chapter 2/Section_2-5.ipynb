##### Objective 2: Use vertical shifts/translations to graph functions

##### Vertical Shifts/Translations

* Let $f$ be a function and $c$ a *positive* real number

    * The graph of $y = f(x) + c$ is the graph of $y = f(x)$ shifted $c$ units **vertically upward**
    
    * The graph of $y = f(x) - c$ is the graph of $y = f(x)$ shifted $c$ units **vertically downward**
    
* In either case, we can begin by graphing the **parent function (or basic function)** and then **transform** the parent function by shifting in the appropriate direction (up or down)

##### Example 5:

Use transformations to graph the function and give the domain and range: $f(x) = x^2 + 4$

##### Solution:

* We note that we have the quadratic function $x^2$ with $4$ added to it

* The graph of $f(x) = x^2 + 4$ can be determined by taking the function $x^2$ and shifting each point *up* by $4$ units

```python
fig, ax = plt.subplots(figsize=(25,15))

x = np.linspace(-10, 10, 100)
y_parent = x**2
y_transf = x**2 + 4
ax.plot(x, y_parent, color='b', ls='--', label=r'$x^2$')
ax.plot(x, y_transf, color='b', label=r'$x^2 + 4$')

ax.axvline(x=0, color='k')
ax.axhline(y=0, color='k')

ax.set_xlim(-10,10)
ax.set_ylim(-10, 10)

ax.set_xticks(np.arange(-10, 11, 1))
ax.set_yticks(np.arange(-10, 11, 1))

ax.tick_params(labelsize=20)
ax.set_title(r'$f(x)$', fontsize=30)
ax.set_aspect('equal')

ax.legend()
```

* Domain: $(-\infty, \infty)$

* Range: $[4, \infty)$

##### Objective 3: Use horizontal shifts/translations to graph functions

##### Horizontal Shifts/Translations

* Let $f$ be a function and $c$ a *positive* real number

    * The graph of $y = f(x + c)$ is the graph of $y = f(x)$ shifted $c$ units **horizontally to the left**
    
    * The graph of $y = f(x - c)$ is the graph of $y = f(x)$ shifted $c$ units **horizontally to the right**

##### Objective 4: Use reflections to graph functions

##### Reflections

* The graph of $y = -f(x)$ is the graph of $y = f(x)$ **reflected about the $x$-axis**

* The graph of $y = f(-x)$ is the graph of $y = f(x)$ **reflected about the $y$-axis**

##### Objective 5: Graph a function involving a sequence of transformations

##### Multiple Transformations

* A function involving more than one transformation can be graphed by performing transformations in the following order:

    1. Reflect (vertically and/or horizontally)
    
    2. Shift/translate (vertically and/or horizontally)
